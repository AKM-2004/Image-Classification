```
import zipfile

zio = zipfile.ZipFile("/content/microsoft-catsvsdogs-dataset.zip",'r')
zio.extractall('/content')
zio.close()
```

```
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

#Now we are generating on single image

```
img  = image.load_img('/content/PetImages/Cat/1.jpg',target_size=(200,200))
```

```
img
```

```
imagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.3,
    horizontal_flip = True,
    vertical_flip = False,
    width_shift_range = 0.3,
    height_shift_range=0.2
)
```

```
img = image.img_to_array(img)
```

```
img.shape
```

```
input = img.reshape(1,200,200,3) # yaha per ek extra dimension add isliye karna pad raha hai kyu ki hume bata na pad ta hai ki hum batch mai kitnai images de rahe hai
```

```
i=0

for output in imagen.flow(input,batch_size = 1,save_to_dir="/content/sample_data"):
  i = i+1
  if i == 10:
    break
```
